## 3) Graphs:

*Reading and opening the data*

```python
import pandas as pd

# reading the data
df_hs = pd.read_csv('C:/Users/ivan2/gitLocal/DSC550-WINTER2023/campus-data-download-4yr-longitudinal-2018-2020-class.csv.csv')
df_hs.head(5)
```

**1)** *Creating a histogram plot to answer the first problem question.*

```python
import matplotlib.pyplot as plt
```

```python
# Create the histogram for Graduation Rates
grad_rates = df_hs.CAMP_ALLR_GRAD
plt.hist(grad_rates, bins=50, label='Graduation Rates')
plt.xlabel('All Schools Graduation Rates')
plt.ylabel('Frequency')
plt.title('Histogram of Graduation Rates')

plt.legend()
plt.show()
```

**2)** *Creating a histogram plot to answer the second problem question.*

```python
# Create the histogram for class size
grad_rates = df_hs.CAMP_ALLD

plt.hist(grad_rates, bins=50, label='Graduation Rates')
plt.xlabel('All Schools Graduation Rates')
plt.ylabel('Frequency')
plt.title('Histogram of Graduation Rates')

plt.legend()
plt.show()
```

**3)** *Creating a pie graph to answer the third problem question.*

```python
# Visualizing Counts the in the Region_Type variable
region_type_counts = df_hs['REGION_TYPE'].value_counts()

pie_labels = region_type_counts.index

plt.figure(figsize=(8, 8))
plt.pie(region_type_counts, labels=region_type_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of High Schools by Region Types')

custom_legend_labels = ['1 = Rural', '2 = Urban']
plt.legend(title="Region Types", labels=custom_legend_labels)

plt.show()

# 1 = Rural
# 2 = Urban
```

## 4) Insights Learned:

These three plots have helped me find answer the problem questions. I learned that histograms are best suited for continuous variables and pie graphs are best suited for categorical variables with low amounts of categories.

- The first histogram shows that graduation rates are left-skwed with most schools having high graduation rates.
- The second histogram shows that there is a large amount of schools with a class size of 0. These are errors and should be filtered out of the dataset before further analysis is conducted.
- The third graph shows that most schools in the state of Texas are located in an Rural region.
