### Reading the cookie training data

```python
# importing libraries
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# reading the data
file_path = 'C:/Users/ivan2/gitLocal/DSC550-WINTER2023/Cookie_training_data.xlsx'


df_ck = pd.read_excel(file_path)
df_ck.head(5)
```

```python
df_ck.shape
```

```python
df_ck['CategoryID'] = df_ck['CategoryID'].astype('category')
df_ck['Category'] = df_ck['Category'].astype('category')
print(df_ck.dtypes)
```

### Conducting initial EDA into the cookie training data

**1. Plotting the counts of the cookie category variable**

```python
# Plotting the counts of the 'CategoryID' column
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Category', data=df_ck)
plt.title('Count of Each Category')
plt.xlabel('Category')
plt.ylabel('Count')


plt.show()
```

**2. Displaying a few cookie descriptions and assigned categories**

```python
def show_descriptions(df_ck, n = 1, RANDOM_STATE = 44):
    temp_df = df_ck.sample(n, random_state = RANDOM_STATE)
    indices = temp_df.index.tolist()
    for i in indices:
        print()
        print("Description:\n")
        print(f"Category: {temp_df.loc[i, 'Category']} \n")
        print(temp_df.loc[i, 'Concat'])
        print()
        print('==' * 10)
```

```python
show_descriptions(df_ck, 3)
```

**3. Plotting distribution of description length**

```python
# establishing a copy of the dataframe
dfV2 = df_ck.copy()
```

```python
# counting the length of each description
dfV2['desc_len'] = dfV2['Concat'].map(len)
# check
dfV2[['Concat', 'desc_len']].head()
```

```python
# function for plotting distribution
def plot_distribution(df, colname, title, bin_size=None):
    sns.displot(data = dfV2, x = colname, bins=bin_size)
    plt.xlabel(title)
    plt.ylabel("Frequency")
    plt.title("Distribution of {}".format(title))
    plt.show();
```

```python
plot_distribution(dfV2, 'desc_len', 'Description Length', bin_size=20)
```

The distribution is unimodal. Most of the descriptions are relatively short, with a few much longer descriptions. The data does not follow a normal distribution and is skewed to the right. This is normal for text data since most entries will be concise.

**4. Plotting distribution of description length per category**

```python
def show_dist_per_cat(df, colname, title):
    sns.boxplot(data = dfV2, y = colname, x = 'Category')
    plt.ylabel("Length")
    plt.xlabel("Category")
    plt.title(f"{title} Distribution per Sentiment")
    plt.show()
    
    print()
    print(dfV2.groupby(['Category'])[colname].describe())
```

```python
show_dist_per_cat(dfV2, 'desc_len', "Description Length")
```

**5. Plotting the counts of the cookie source variable**

```python
# Sorting the value counts
sorted_source_counts = dfV2['Source'].value_counts().sort_values(ascending=False)
```

```python
# Plotting the counts of the 'Source' column
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Source', data=dfV2, order=sorted_source_counts.index)
plt.title('Count Sources')
plt.xlabel('Sources')
plt.ylabel('Count')

# Rotate x-axis labels to vertical
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

plt.show()
```

**EDA Summary**

In short, the training data consists of a high volume of Essential Cookies and a low volume of Do Not Sell / Do Not Share Cookies. I will have to be careful when splitting the data into training and test sets to ensure that the low-volume cookies are adequately represented in each set.

The distribution of description length is right skewed and unimodal, which can be normal for text-based data.

Essential and DNS/DNS cookies' average description length tends to be high, while Non-Essential cookies tend to have smaller descriptions. There are a few outliers with significantly long descriptions. I will consider removing these outliers during the modeling phase.

The sources with the highest cookie counts are from our web analytics platforms. This is expected since those are first-party cookies. Ultimately, the exploratory data analysis gives me a good understanding of the data and leads me down a path for my next steps.
